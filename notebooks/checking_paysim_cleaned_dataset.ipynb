```python
# in a notebook or script cell
import pandas as pd
df = pd.read_csv("cleaned_paysim.csv")
print(df[['upi_type_upi_payment','upi_type_upi_transfer']].head())
print(df[['upi_type_upi_payment','upi_type_upi_transfer']].dtypes)

# If they are object / bool, convert to numeric:
df['upi_type_upi_payment'] = pd.to_numeric(df['upi_type_upi_payment'], errors='coerce').fillna(0).astype(int)
df['upi_type_upi_transfer'] = pd.to_numeric(df['upi_type_upi_transfer'], errors='coerce').fillna(0).astype(int)

print("After convert:", df[['upi_type_upi_payment','upi_type_upi_transfer']].dtypes)
```
